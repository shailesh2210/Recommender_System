```python
new_meta["id"] = new_meta["id"].astype(int)
```

```python
data = pd.merge(new_meta, credits, on="id")
```

```python
data.head()
```

```python
data["genres"].dtype
```

```python
import ast
data["cast"] = data["cast"].map(lambda x: ast.literal_eval(x))
data["crew"] = data["crew"].map(lambda x: ast.literal_eval(x))
data["genres"] = data["genres"].map(lambda x: ast.literal_eval(x))
```

```python
def make_genres(x):
    gen = []
    st = " "
    for i in x:
        if i.get("name")=="Science Fiction":
            scifi = "scifi"
            gen.append(scifi)
        else:
            gen.append(i.get("name"))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
            
```

```python
data["genres_list"] = data["genres"].map(lambda x: make_genres(x))
```

```python
data["genres_list"]
```

```python
def get_actor(x):
    cast = []
    for i in x:
        cast.append(i.get("name"))
    if cast == []:
        return np.NaN
    else:
        return (cast[0])
```

```python
data["actor_1_name"] = data["cast"].map(lambda x: get_actor(x))
```

```python
data["actor_1_name"]
```

```python
def get_actor2(x):
    cast = []
    for i in x:
        cast.append(i.get("name"))
    if cast == [] or len(cast)<=1:
        return np.NaN
    else:
        return (cast[1])
```

```python
data["actor_2_name"] = data["cast"].map(lambda x: get_actor2(x))
```

```python
data["actor_2_name"]
```

```python
def get_actor3(x):
    cast = []
    for i in x:
        cast.append(i.get("name"))
    if cast == [] or len(cast)<=2:
        return np.NaN
    else:
        return (cast[2])
```

```python
data["actor_3_name"] = data["cast"].map(lambda x: get_actor3(x))
```

```python
data["actor_3_name"]
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get("job") == "Director":
            dt.append(i.get("name"))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data["director_name"] = data["crew"].map(lambda x: get_directors(x))
```

```python
data["director_name"]
```

```python
data.columns
```

```python
movies = data.loc[:,["director_name",'actor_1_name', 'actor_2_name', 'actor_3_name','genres_list','title' ]]
```

```python
movies.head()
```

```python
movies = movies.rename(columns={"genres_list":"genres"})
movies = movies.rename(columns={"title":"movie_title"})
```

```python
movies.isna().sum()
```

```python
movies = movies.dropna(how="any")
```

```python
movies["movie_title"] = movies["movie_title"].str.lower()
```

```python
movies["comb"] = movies["actor_1_name"] + " " + movies["actor_2_name"]+ " " + movies["actor_3_name"]+ " " + movies["director_name"]+ " "+ movies["genres"]
```
